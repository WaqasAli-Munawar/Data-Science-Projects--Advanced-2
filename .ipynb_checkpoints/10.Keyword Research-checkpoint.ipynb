Google Trends are a keyword research tool that helps the researchers, bloggers, digital marketers and some more people in the digital industry to find how often a keyword is entered into Google search engine over a given period. Google Trends is used for keyword research mostly is writing articles on hot topics. 

In this file, we will perform keyword research to find the hottest topics and keywords.

We will use the **Google API** to access Google trends which can be done by using the **pytrends** library in python. Python being a general-purpose programming language provides libraries and packages for almost every task. **pytrends** can be easily installed by using the **pip** command – `pip install pytrends`.

### Keyword Research with Python

We need to log in to Google first because, after all, we ask Google Trends for trending topics. For that, we need to import the method called **TrendReq from the pytrends.request** method:

```python
import pandas as pd
from pytrends.request import TrendReq
import matplotlib.pyplot as plt
```

```python
trends = TrendReq()
```

### Keyword Interest By Region
Let’s see the terms that are popular in the region around the world. We will choose the term to search for as **Data Science**:

```python
trends.build_payload(kw_list=["Data Science"])
data = trends.interest_by_region()
```

```python
data.sort_values("Data Science", ascending = False)
```

```python
data.sample(10)
```

Values are calculated on a scale of **0 to 100**, where 
* **100** is the most popular location as a fraction of the total searches for that location, 
* a value of **50** indicates a location half as popular. 

Now let’s visualize the above search results to get better insights

```python
import matplotlib.style as style

print(style.available)
```

```python
df = data.sample(15)
# style.use("default")
df.reset_index().plot(x="geoName", y="Data Science", figsize=(120,16), kind="bar")
plt.show()
```

### Keyword Research for Daily Search Trends
Now let’s take a look at the top daily search trends around the world. To do this, we need to use the `trending_searches()` method:

```python
data = trends.trending_searches(pn="united_states")
print(data.head(10))
```

```python
data = trends.trending_searches(pn="canada")
print(data.head(10))
```

### Google Keyword Suggestion

Now, let’s see how we can get the google keywords suggestion for keyword research. Keywords are those words that are mostly typed by users in the search engine to find about a particular topic:

```python
keyword = trends.suggestions(keyword="Programming")
data = pd.DataFrame(keyword)
data.head()
```
